```python
announced_df = pd.read_csv("spac-research.csv", encoding = "ISO-8859-1")
```

```python
announced_df.head()
```

```python
badHeaders = []

for header in list(announced_df.columns.values):
    if (header != "Ticker" and header != "Announced" and header != "Closed"):
        badHeaders.append(header)

announced_df = announced_df.drop(badHeaders, axis=1)
```

```python
announced_df.to_csv('filtered_announced_data.csv')
```

```python
min_df = pd.read_excel("mindata2.xlsx", engine="openpyxl", encoding = "ISO-8859-1")
```

```python
cols = min_df.columns.values
badCols = []
for col in cols:
    if (min_df[col][1] != min_df[col][1]):
        badCols.append(col)
min_df = min_df.drop(badCols, axis=1)
min_df
```

```python
cols = min_df.columns.values
print(len(cols))
for x in range(len(cols) / 2):
    name = cols[x * 2]
    cols[x * 2] = "Date"
    cols[x * 2 + 1] = name.split()[0]
    min_df.columns = cols   
        
min_df
```
