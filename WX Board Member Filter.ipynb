```python
import pandas as pd
```

```python
def uniq(test_list):
    res = []
    for i in test_list:
        if i not in res:
            res.append(i)
    return test_list
```

```python
df= pd.read_csv('spac-research.csv')
df.head()
```

```python
df[['President','CEO','Chairman']]
```

```python
df['column_name'] = df['President']+ ' ' + df['CEO'] # + df['Director'] + ....
```

```python
people = list(df['CEO'])
people.append(list(df['Chairman']))
# have another dozen statements appending a list of the names of directors and stuff
###############
people = uniq(people)
len(people), people
# removing the nan and removing the suffixes like MBA and PHD
people = [x.split(',')[0] for x in people if type(x) == str]
len(people), people
```
